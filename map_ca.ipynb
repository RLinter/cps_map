```python
# prompt: draw map of California with no other states using geopandas.

!pip install geopandas
import geopandas as gpd
import matplotlib.pyplot as plt

# Download the shapefile for US states (you might need to adjust the URL if needed)
!wget https://www2.census.gov/geo/tiger/GENZ2018/shp/cb_2018_us_state_500k.zip
!unzip cb_2018_us_state_500k.zip

# Read the shapefile into a GeoDataFrame
states = gpd.read_file("cb_2018_us_state_500k.shp")

# Filter for Califronia
california = states[states['NAME'] == 'California']

# Create the plot
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
california.plot(ax=ax, color='lightblue', edgecolor='black')

# Set title and remove axis
ax.set_title('California')
ax.axis('off')

# Show the plot
plt.show()
```

```python
# prompt: draw map of California with no other states using geopandas. Also draw each county borders in the map.

import geopandas as gpd
import matplotlib.pyplot as plt

# Download the shapefile for US counties (you might need to adjust the URL if needed)
!wget https://www2.census.gov/geo/tiger/GENZ2018/shp/cb_2018_us_county_500k.zip
!unzip cb_2018_us_county_500k.zip

# Read the shapefile into a GeoDataFrame
counties = gpd.read_file("cb_2018_us_county_500k.shp")

# Filter for California counties
california_counties = counties[counties['STATEFP'] == '06']

# Create the plot
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

# Plot Florida counties
california_counties.plot(ax=ax, color='lightblue', edgecolor='black', linewidth=0.5)

# Set title and remove axis
ax.set_title('California Counties')
ax.axis('off')

# Show the plot
plt.show()
```
